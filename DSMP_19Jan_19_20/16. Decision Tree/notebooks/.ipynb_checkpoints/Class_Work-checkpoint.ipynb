# Read Data

```python
import pandas as pd
import numpy as np
films = pd.read_csv('../data/films.csv')
films.head()
```

```python
films.gender.value_counts()
films['is_28+'].value_counts()
films.employment_status.value_counts()
films.watching.value_counts()
```

```python
films = pd.get_dummies(films)
films.rename({'is_28+': 'is_greater_28'}, axis=1, inplace=True)
films = films[['is_greater_28', 'gender_F', 'employment_status_student', 'watching_yes']]
films.head()
```

```python
X = films.iloc[:, :-1]
y = films.iloc[:, -1]

feature_cols = X.columns

X.head()
y[:5]
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X, y)
```

```python
dot_data = StringIO()
export_graphviz(clf, out_file=dot_data,  
                filled=True, rounded=True,
                special_characters=True, feature_names = feature_cols,class_names=['0','1'])

graph = pydotplus.graph_from_dot_data(dot_data.getvalue())  
#graph.write_png('diabetes.png')
Image(graph.create_png(), width = 600, height = 300)
```
